1. PENGGUNAAN NUMPY

```python
import numpy as np
```

```python
# Buat array
arr = np.array([1, 2, 3, 4, 5])
print(arr)
print(arr.shape)
print(arr.dtype)
```

```python
# Operasi matematika — berlaku ke semua elemen sekaligus
print(arr * 2)
print(arr + 10)
print(arr ** 2)
```

```python
# Array 2D (matrix)
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print(matrix.shape)  # (2, 3) → 2 baris, 3 kolom
```

```python
# Slicing
print(arr[1:4])       # index 1 sampai 3
print(matrix[0, :])   # baris pertama, semua kolom
```

2. PENGGUNAAN PANDAS

```python
import pandas as pd
```

```python
# Buat DataFrame manual
data = {
    'nama': ['Andi', 'Budi', 'Citra', 'Dian'],
    'usia': [25, 30, 22, 28],
    'gaji': [5000000, 8000000, 4500000, 7000000]
}
```

```python
df = pd.DataFrame(data)
print(df)
print(df.shape)        # jumlah baris dan kolom
print(df.dtypes)       # tipe data tiap kolom
print(df.describe())   # statistik dasar
```

```python
# Filtering
print(df[df['usia'] > 25])           # filter baris
print(df[['nama', 'gaji']])          # pilih kolom tertentu
```

```python
# Tambah kolom baru
df['gaji_juta'] = df['gaji'] / 1_000_000
print(df)
```

```python
# Sorting
print(df.sort_values('gaji', ascending=False))
```
